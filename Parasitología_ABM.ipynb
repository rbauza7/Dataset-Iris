```r
# Instalar paquetes si es necesario
install.packages("dplyr")

# Cargar librería
library(dplyr)

# Seleccionar las columnas de los parásitos
parasite_cols <- c("Al", "Complejo", "Ec", "En", "Bsp", "Gl", "Ib", "Chm", "Ph")

# Calcular la frecuencia de infección por cada parásito
parasite_counts <- datos %>%
  select(all_of(parasite_cols)) %>%
  summarise_all(~sum(. == 1, na.rm = TRUE))

# Mostrar el resultado
print(parasite_counts)
```

```r
# Instalar paquete si es necesario
install.packages("ggplot2")

# Cargar librerías
library(ggplot2)
library(tidyr)

# Transformar datos para gráfico
df_parasitos <- gather(parasite_counts, key = "Parásito", value = "Casos")

# Crear gráfico de barras
ggplot(df_parasitos, aes(x = Parásito, y = Casos, fill = Parásito)) +
  geom_bar(stat = "identity") +
  theme_minimal() +
  labs(title = "Distribución de parásitos en la muestra",
       x = "Tipo de parásito", y = "Número de casos") +
  theme(legend.position = "none")  # Ocultar leyenda innecesaria
```

```r
# Crear una función para hacer Chi-cuadrado entre Procedencia y cada parásito
chi_test_parasito <- function(parasito) {
  tabla <- table(datos[[parasito]], datos$Procedencia)
  test <- chisq.test(tabla)
  return(data.frame(Parasito = parasito, P_Value = test$p.value))
}

# Lista de parásitos
parasite_cols <- c("Al", "Complejo", "Ec", "En", "Bsp", "Gl", "Ib", "Chm", "Ph")

# Aplicar la función a cada parásito
resultados_chi <- do.call(rbind, lapply(parasite_cols, chi_test_parasito))

# Mostrar los resultados ordenados por significancia
resultados_chi <- resultados_chi[order(resultados_chi$P_Value), ]
print(resultados_chi)
```
